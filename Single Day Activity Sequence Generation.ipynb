```python
from sklearn import preprocessing
import pandas as pd
import datetime
```

```python
df_File = pd.read_csv("Datasets\\r4.2\\file.csv" ,usecols= ['date', 'user','filename'])
df_Device = pd.read_csv("Datasets\\r4.2\\device.csv" ,usecols= ['date', 'user','pc','activity'])
df_Logon = pd.read_csv("Datasets\\r4.2\\logon.csv" ,usecols= ['date', 'user','pc','activity'])
df_Email = pd.read_csv("Datasets\\r4.2\\email.csv" ,usecols= ['date', 'user','to','cc','bcc','size','attachments'])
```

```python
logon_data = df_Logon

user_data = []

unique_users = logon_data.user.unique().tolist()

for user in unique_users:
    user_logon_data = logon_data[logon_data['user'] == user].sort_values('date')

    date_list = pd.to_datetime(user_logon_data['date']).dt.date
    user_logon_data['logon_date'] = date_list

    grouped_logon_data = user_logon_data.groupby(['logon_date'])

    for date, details in grouped_logon_data:
        start_time = datetime.datetime(year=date.year, month=date.month, day=date.day, hour=8, minute=0, second=0)
        end_time = datetime.datetime(year=date.year, month=date.month, day=date.day, hour=19, minute=0, second=0)

        date_logon_list = details.date.tolist()
        activity_list = details.activity.tolist()

        start_count = 0
        end_count = 0
        total_time_start = 0
        total_time_end = 0
        session_duration_start = 0
        session_duration_after_office_hours = 0

        for logon_time, log_activity in zip(date_logon_list, activity_list):
            logon_time = datetime.datetime.strptime(logon_time, "%m/%d/%Y %H:%M:%S")

            if logon_time > end_time and log_activity == 'Logon':
                time_since_start = (logon_time - start_time).total_seconds() / 3600
                start_count += 1
                time_since_end = (logon_time - end_time).total_seconds() / 3600
                end_count += 1

                index = activity_list.index(log_activity)
                if index < len(activity_list):
                    if index + 1 <= len(activity_list):
                        if activity_list[index + 1] == 'Logoff':
                            logoff_time = datetime.datetime.strptime(date_logon_list[index + 1], "%m/%d/%Y %H:%M:%S")
                            session_duration_after_office_hours += ((logon_time - logoff_time).total_seconds()) / 3600

        avg_start_time = 0
        if start_count > 0:
            avg_start_time = time_since_start / start_count

        avg_end_time = 0
        if end_count > 0:
            avg_end_time = time_since_end / end_count

        logon_indices = [i for i, activity in enumerate(activity_list) if activity == "Logon"]       
        num_of_logins = len(logon_indices)

        if num_of_logins > 1:
            first_login_index = logon_indices[0]
            first_login_time = datetime.datetime.strptime(date_logon_list[first_login_index], "%m/%d/%Y %H:%M:%S")
            first_login_diff = (first_login_time - start_time).total_seconds() / 3600 if first_login_time > start_time else (start_time - first_login_time).total_seconds() / 3600

        if num_of_logins > 1:
             last_login_index = logon_indices[-1]
             last_login_time = datetime.datetime.strptime(date_logon_list[last_login_index], "%m/%d/%Y %H:%M:%S")
             last_login_diff = (last_login_time - start_time).total_seconds() / 3600 if last_login_time > start_time else (start_time - last_login_time).total_seconds() / 3600

        pc_list = details.pc.tolist()
        num_of_unique_pcs_accessed = len(set(pc_list))

        pcs_after_office = []
        session_count_after_office = 0
        for index in logon_indices:
            if datetime.datetime.strptime(date_logon_list[index], "%m/%d/%Y %H:%M:%S") > end_time:
                session_count_after_office += 1
                pcs_after_office.append(pc_list[index])

        user_features = {}
        user_features['User'] = user
        user_features['Date'] = date
        user_features['First_Login_Diff'] = first_login_diff
        user_features['Last_Login_Diff'] = last_login_diff
        user_features['Avg_Start_Time'] = avg_start_time
        user_features['Avg_End_Time'] = avg_end_time
        user_features['Num_Of_Logins'] = num_of_logins
        user_features['Session_Count_After_Office'] = session_count_after_office
        user_features['Num_Of_Unique_PCs_Accessed'] = num_of_unique_pcs_accessed
        user_features['Num_Of_PCs_After_Office'] = len(pcs_after_office)
        user_features['Session_Duration_After_Office_Hours'] = session_duration_after_office_hours

        user_data.append(user_features)     

logon_features_df = pd.DataFrame(user_data) 
print(logon_features_df.head())
```

```python
device_data = df_Device

user_list = device_data.user.unique().tolist()
device_info = []

for usr in user_list:
    user_device_data = device_data[device_data['user'] == usr].sort_values('date')

    date_list = pd.to_datetime(user_device_data['date']).dt.date
    user_device_data['device_date'] = date_list

    grouped_device_data = user_device_data.groupby(['device_date']) 

    for dt, det in grouped_device_data:
        device_activity_list = det.activity.tolist()       
        connect_indices = [i for i, act in enumerate(device_activity_list) if act == "Connect"]
        D2_device_usage_count = len(connect_indices)

        D1_connect_after_office = 0
        if D2_device_usage_count > 0:
            device_time_list = det.date.tolist()
            office_end = datetime.datetime(year=dt.year, month=dt.month, day=dt.day, hour=19, minute=0, second=0)

            for ind in connect_indices:
                if datetime.datetime.strptime(device_time_list[ind], "%m/%d/%Y %H:%M:%S") > office_end:
                    D1_connect_after_office += 1              

        user_device_details = {}
        user_device_details['User'] = usr
        user_device_details['Date'] = dt
        user_device_details['Devices_Connected_After_Office'] = D1_connect_after_office
        user_device_details['Total_Device_Usage_Count'] = D2_device_usage_count

        device_info.append(user_device_details)

device_features_df = pd.DataFrame(device_info)
print(device_features_df.head())
```

```python
file_data = df_File

users = file_data.user.unique().tolist()
file_info = []

for user in users:
    user_data = file_data[file_data['user'] == user].sort_values('date')

    date_list = pd.to_datetime(user_data['date']).dt.date
    user_data['file_date'] = date_list

    grouped_data = user_data.groupby(['file_date']) 

    for dt, det in grouped_data:
        file_names = det.filename.tolist()
        exe_download_count = 0
        others_download_count = 0
        for file_name in file_names:
            if file_name.endswith(".exe"):
                exe_download_count += 1
            else:
                others_download_count +=1
        file_details = {}
        file_details['User'] = user
        file_details['Date'] = dt
        file_details['EXE File Count'] = exe_download_count
        file_details['Other File Count'] = others_download_count
        file_info.append(file_details)                 

file_features_df = pd.DataFrame(file_info) 
print(file_features_df.head())
```

```python
email_data = df_Email

unique_users = email_data.user.unique().tolist()
email_features = []

for user in unique_users:
    user_email_data = email_data[email_data['user'] == user].sort_values('date')

    date_list = pd.to_datetime(user_email_data['date']).dt.date
    user_email_data['email_date'] = date_list

    grouped_email_data = user_email_data.groupby(['email_date']) 

    for email_date, email_details in grouped_email_data:
        total_size = user_email_data.loc[user_email_data['email_date'] == email_date, 'size'].sum()
        to_list = email_details.to.tolist()
        cc_list = email_details.cc.tolist()
        bcc_list = email_details.bcc.tolist()
        email_size = email_details.size       
        avg_size_ratio = 0
        avg_size_ratio = email_size / total_size
        attachments_list = email_details.attachments.tolist()
        attachments_list =  [attachment for attachment in attachments_list if str(attachment) != 'nan']
        attachment_count = user_email_data.loc[user_email_data['email_date'] == email_date, 'attachments'].sum()       
        recipients_list = to_list + cc_list + bcc_list
        recipients_list = [recipient for recipient in recipients_list if str(recipient) != 'nan']
        to_list =  [to for to in to_list if str(to) != 'nan']
        cc_list =  [cc for cc in cc_list if str(cc) != 'nan']
        bcc_list = [bcc for bcc in bcc_list if str(bcc) != 'nan']    
        total_recipients = 0
        total_recipients = len(to_list) + len(cc_list) + len(bcc_list)
        to_list = ';'.join(to_list)
        cc_list = ';'.join(cc_list)
        bcc_list = ';'.join(bcc_list)    
        recipients_list = to_list + cc_list + bcc_list
        recipients_list = str(to_list).split(";")    
        outside_recipient_count = 0
        for email_id in recipients_list:
            if not email_id.endswith("@dtaa.com"):
                outside_recipient_count = outside_recipient_count + 1
        email_info = {}
        email_info['User'] = user
        email_info['Date'] = email_date
        email_info['Outside_Recipients'] = outside_recipient_count
        email_info['Attachment_Count'] = attachment_count
        email_info['Avg_Size_Ratio'] = avg_size_ratio
        email_info['Total_Recipients'] = total_recipients

        email_features.append(email_info)     

email_features_df = pd.DataFrame(email_features)
print(email_features_df.head())
```

```python
new_df = pd.merge(logon_features_df, email_features_df,  on=['Date','User'] , how = 'outer')    
#new_df = pd.merge(new_df, df_Http,  on=['Date','User'] , how = 'outer')    
new_df = pd.merge(new_df, file_features_df,  on=['Date','User'] , how = 'outer')    
new_df = pd.merge(new_df, device_features_df,  on=['Date','User'] , how = 'outer')    
new_df["First_Login_Diff"].fillna("0", inplace = True) 
new_df["Last_Login_Diff"].fillna("0", inplace = True) 
new_df["Avg_Start_Time"].fillna("0", inplace = True) 
new_df["Avg_End_Time"].fillna("0", inplace = True) 
new_df["Num_Of_Logins"].fillna("0", inplace = True) 
new_df["Session_Count_After_Office"].fillna("0", inplace = True) 
new_df["Num_Of_Unique_PCs_Accessed"].fillna("0", inplace = True) 
new_df["Num_Of_PCs_After_Office"].fillna("0", inplace = True) 
new_df["Session_Duration_After_Office_Hours"].fillna("0", inplace = True)     
new_df["Outside_Recipients"].fillna("0", inplace = True) 
new_df["Attachment_Count"].fillna("0", inplace = True) 
new_df["Avg_Size_Ratio"].fillna("0", inplace = True) 
new_df["Total_Recipients"].fillna("0", inplace = True)    
new_df["EXE File Count"].fillna("0", inplace = True) 
new_df["Other File Count"].fillna("0", inplace = True) 
new_df["Devices_Connected_After_Office"].fillna("0", inplace = True) 
new_df["Total_Device_Usage_Count"].fillna("0", inplace = True)     

print(new_df.head())
```

```python
def classify_scenario(row):
    if float(row["Session_Duration_After_Office_Hours"]) > 0 and \
       int(row["Num_Of_Unique_PCs_Accessed"]) > 0 and \
       int(row["Num_Of_Unique_PCs_Accessed"]) < 3 and \
       int(row["Num_Of_PCs_After_Office"]) > 0 and \
       int(row["Num_Of_PCs_After_Office"]) < 3 and \
       int(row["Outside_Recipients"]) > 0 and \
       int(row["Attachment_Count"]) > 0 and \
       int(row["Attachment_Count"]) < 3 and \
       int(row["Total_Recipients"]) > 50:
        return "Scenario S"
    elif float(row["Avg_Start_Time"]) == 0 and \
         float(row["Avg_End_Time"]) == 0 and \
         int(row["Session_Count_After_Office"]) == 0 and \
         float(row["Session_Duration_After_Office_Hours"]) > 0 and \
         int(row["Num_Of_PCs_After_Office"]) > 0 and \
         int(row["Num_Of_PCs_After_Office"]) > int(row["Num_Of_Unique_PCs_Accessed"]):
        return "Scenario S2"
    elif int(row["Num_Of_Logins"]) > 0 and \
         int(row["Num_Of_Logins"]) < 4 and \
         int(row["EXE File Count"]) > 0 and \
         int(row["Other File Count"]) > 0 and \
         int(row["Devices_Connected_After_Office"]) > 0 and \
         int(row["Total_Device_Usage_Count"]) > 0:
        return "Scenario S3"
    elif int(row["Num_Of_Logins"]) > 10 and \
         int(row["Num_Of_Logins"]) < 90 and \
         int(row["Num_Of_Unique_PCs_Accessed"]) > 0 and \
         int(row["Num_Of_Unique_PCs_Accessed"]) < 10 and \
         int(row["Total_Recipients"]) > 0 and \
         int(row["Attachment_Count"]) > 0 and \
         int(row["Attachment_Count"]) < 10:
        return "Scenario S4"
    elif float(row["Avg_Size_Ratio"]) > 0 and \
         float(row["Avg_Size_Ratio"]) < 0.5 and \
         int(row["Total_Recipients"]) > 5:
        return "Scenario S5"
    else:
        return "-"

new_df["Class_Label"] = new_df.apply(classify_scenario, axis=1)

print(new_df.head())
```

```python
logon = pd.read_csv("Datasets\\r4.2\\logon.csv")
device = pd.read_csv("Datasets\\r4.2\\device.csv")
file = pd.read_csv("Datasets\\r4.2\\file.csv")
email = pd.read_csv("Datasets\\r4.2\\email.csv")

# Sort and aggregate action IDs based on the date of the action sequence
def sort_by_action_date(data):
    sorted_by_date = []

    for user, user_data in data.groupby('user'):
        user_actions = user_data.sort_values('date')
        sorted_by_date.append({'User': user, 'Sorted_By_Date_Action_IDs': user_actions['id'].tolist()})

    return sorted_by_date

# Call the function for each dataset
sorted_by_date_logon_actions = sort_by_action_date(logon)
sorted_by_date_device_actions = sort_by_action_date(device)
sorted_by_date_email_actions = sort_by_action_date(email)
sorted_by_date_file_actions = sort_by_action_date(file)

# Combine all sorted action IDs into one list for each user
all_sorted_by_date_actions = (
    sorted_by_date_logon_actions +
    sorted_by_date_device_actions +
    sorted_by_date_email_actions +
    sorted_by_date_file_actions
)

# Create a DataFrame with sorted action ID lists for each user based on date
user_sorted_by_date_actionID_list = pd.DataFrame(all_sorted_by_date_actions)

new_df = pd.merge(new_df, user_sorted_by_date_actionID_list, on='User', how='outer')

print(new_df.head())
```
